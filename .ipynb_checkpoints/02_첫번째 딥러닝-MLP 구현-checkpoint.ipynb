##  

## network :    

```python
# ()
model =keras.Sequential()

#  layer  
model.add(keras.layers.InputLayer((28,28))) # layer input layer( shape)
model.add(keras.layers.Flatten()) #28,28  1   (784,)  
model.add(keras.layers.Dense(units=128, activation ='relu'))
model.add(keras.layers.Dense(units=64, activation='relu'))
model.add(keras.layers.Dense(units=10, activation='softmax'))
```

```python
a =np.array([[1,2],[3,4]])
a.shape
```

```python
a.flatten()
```

##  
-      
- 
- Optimizer( )
- 

```python
model.compile(optimizer="adam", # -  (wight) 
             loss ="categorical_crossentropy", #Loss(,)  ,   
             metrics=['accuracy']
             )
```

##  
- X (Input Data Image)
    - 0 ~ 1    . (Scaling)
- y (Output Data)
    - one hot encoding 
        - Label   One Hot Encoding .
    - tensorflow.keras  to_categorical()  

```python
train_image.min(), train_image.max(), train_image.dtype
```

```python
X_train=train_image/255.0  # MinMax     
X_train.dtype
```

```python
X_train=(train_image/255.0).astype(np.float32) # 32    ..  32   
X_train.dtype
```

```python
X_test =(test_image/255.0).astype(np.float32)
```

```python
np.min(X_train), np.max(X_train), X_train.dtype
```

```python
train_label, np.unique(train_label)
```

```python
#Label onehot encoding
y_train = keras.utils.to_categorical(train_label) 
y_test= keras.utils.to_categorical(test_label)

 # onehot encoding         .   .  unit 10 .
```

```python
print(y_train.shape, train_label.shape)
y_train[:5]
```

##  (fit)

```python
# model.fit() :History  
# History : Train      
```

```python
# 1 epoch: train dataset     .
# 1 step :  (weight)  update  
hist=model.fit(X_train,y_train, # train dataset -x,y
         epochs =10,  # epoch  - trian dataset    ,   10 
         batch_size=100, # (weight-  )    .
         validation_split=0.3  #tain set 30%    .
         )

#  : 60000 , batch_size:100
# 1 epoch 600 step   batch_size  100   6     600    600 step  
#  42000   step 420   batch_size  : 100   6 70% 42000   
#Epoch 4/10
#420/420 [==============================] - 2s 4ms/step - loss: 0.0743 - accuracy: 0.9771 - val_loss: 0.1212 - val_accuracy: 0.9638
# 420
```

```python
hist.history['loss'] #     . 
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,20))

plt.subplot(2,1,1)
plt.plot(range(1,11), hist.history['loss'], label='Train Loss' )
plt.plot(range(1,11), hist.history['val_loss'], label='Validation Loss')
plt.title('Loss per Epoch', fontsize=20)
plt.legend()

plt.subplot(2,1,2)
plt.plot(range(1,11), hist.history['accuracy'], label ='Train Accuracy')
plt.plot(range(1,11), hist.history['val_accuracy'], label= 'Validation Accuray')
plt.title("Accuracy per Epoch", fontsize=20)

plt.legend()
plt.tight_layout()
plt.show()
```

##  

```python
test_loss, test_acc = model.evaluate(X_test, y_test) #(X,y).  :Loss [, metrics  ]
```
